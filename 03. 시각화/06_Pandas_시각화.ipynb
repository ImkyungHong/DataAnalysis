```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='NanumBarunGothic')
```

```python
iris = sns.load_dataset("iris")    # 붓꽃 데이터
titanic = sns.load_dataset("titanic")    # 타이타닉호 데이터
tips = sns.load_dataset("tips")    # 팁 데이터
flights = sns.load_dataset("flights")    # 여객운송 데이터
```

## 예시

```python
np.random.seed(2021)
df1 = pd.DataFrame(np.random.randn(100, 3),
                   index=pd.date_range('1/1/2021', periods=100),
                   columns=['A', 'B', 'C']).cumsum()
df1.tail()
```

```python
df1.plot()
plt.show()
```

```python
df1.plot()
plt.title('Pandas의 plot method 사용 예')
plt.xlabel('일자')
plt.ylabel('데이터')
plt.show()
```

## Bar chart

```python
iris.sepal_length[:20].plot(kind='bar', rot=0)
plt.title("꽃받침의 길이 시각화")
plt.xlabel("Data")
plt.ylabel("꽃받침의 길이")
plt.show()      #현재 pandas에서는 컬러 단일화로 나옴. 
```

```python
iris.sepal_length[:20].plot(kind='bar', rot=0, #barh 전환
                            color=plt.get_cmap('RdBu')(np.linspace(0.8,0.2,20)))
plt.title("꽃받침의 길이 시각화")
plt.xlabel("Data")
plt.ylabel("꽃받침의 길이")
plt.show()      # 다양한 컬러 나오게 하기 위해서는 get_cmap 사용 
```

```python
iris.sepal_length[:20].plot(kind='barh', rot=0, #barh 전환
                            color=plt.get_cmap('RdBu')(np.linspace(0.8,0.2,20)))
plt.title("꽃받침의 길이 시각화")
plt.ylabel("Data")  #barh 하면서 이름도 수정해줘야해 
plt.xlabel("꽃받침의 길이")
plt.show()      # 다양한 컬러 나오게 하기 위해서는 get_cmap 사용 
```

```python
iris.sepal_length[:20].sort_values().plot(kind='barh', rot=0, #barh 전환
                            color=plt.get_cmap('RdBu')(np.linspace(0.8,0.2,20)))
plt.title("꽃받침의 길이 시각화")
plt.ylabel("Data")
plt.xlabel("꽃받침의 길이")
plt.show()      # 다양한 컬러 나오게 하기 위해서는 get_cmap 사용 
```

```python
df2 = iris.groupby(iris.species).mean()
df2.columns.name = 'feature'
df2
```

```python
df2.plot.bar(rot=0)
plt.title('각 종의 feature별 평균')
plt.ylim(0,8)
plt.show()
```

```python
df2.T.plot.bar(rot=0)
plt.title('각 feature의 종별 평균')
plt.ylim(0,8)
plt.show()
```

## Pie chart

```python
df3 = titanic['class'].value_counts()
df3.plot.pie(autopct='%.2f%%')
plt.title('선실별 승객 수 비율')
plt.axis('equal')
plt.show()
```
