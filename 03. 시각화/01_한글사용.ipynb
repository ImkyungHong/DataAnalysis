## 2) Colab 런타임 반드시 다시 시작

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# minus 표시
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='NanumBarunGothic')
```

```python
plt.title('플롯 예')
plt.plot([0,1,4,9])    
plt.show()
```
